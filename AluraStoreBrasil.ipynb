#1. Análise do faturamento



```python
def calcular_faturamento_lojas(*lojas):
    faturamento_total = []
    for loja_df in lojas:
        faturamento_total.append(loja_df['Preço'].sum())
    return faturamento_total
```

```python
faturamento = []

faturamento.append(loja['Preço'].sum())
faturamento.append(loja2['Preço'].sum())
faturamento.append(loja3['Preço'].sum())
faturamento.append(loja4['Preço'].sum())

for i in range(len(faturamento)):
  print(f'Faturmento loja {i+1}: {faturamento[i]}\n')
```

# 2. Vendas por Categoria

```python
categoria_loja = loja['Categoria do Produto'].value_counts()
categoria_loja2 = loja2['Categoria do Produto'].value_counts()
categoria_loja3 = loja3['Categoria do Produto'].value_counts()
categoria_loja4 = loja4['Categoria do Produto'].value_counts()
```

```python
print('Categoria loja 1:')
print(categoria_loja)
print('\nCategoria loja 2:')
print(categoria_loja2)
print('\nCategoria loja 3:')
print(categoria_loja3)
print('\nCategoria loja 4:')
print(categoria_loja4)
```

# 3. Média de Avaliação das Lojas


```python
media_de_avaliacao_loja1 = loja['Avaliação da compra'].mean()
media_de_avaliacao_loja2 = loja2['Avaliação da compra'].mean()
media_de_avaliacao_loja3 = loja3['Avaliação da compra'].mean()
media_de_avaliacao_loja4 = loja4['Avaliação da compra'].mean()
```

```python
medias_avaliacoes = [
    media_de_avaliacao_loja1,
    media_de_avaliacao_loja2,
    media_de_avaliacao_loja3,
    media_de_avaliacao_loja4
]

for i, media in enumerate(medias_avaliacoes):
  print(f'Média de avaliação loja {i+1}: {media:.2f}')
```

# 4. Produtos Mais e Menos Vendidos

```python
def mais_e_menos_vendido(df, coluna = 'Produto'):
    contagem = df[coluna].value_counts()

    mais = (contagem.idxmax(), contagem.max())
    menos = (contagem.idxmin(), contagem.min())

    print(f'O mais vendido foi: {mais}')
    print(f'O menos vendido foi: {menos}')
    return [mais, menos]
```

```python
mais_e_menos_vendido(loja, coluna="Produto")
```

```python
mais_e_menos_vendido(loja2, coluna="Produto")
```

```python
mais_e_menos_vendido(loja3, coluna="Produto")
```

```python
mais_e_menos_vendido(loja4, coluna="Produto")
```
